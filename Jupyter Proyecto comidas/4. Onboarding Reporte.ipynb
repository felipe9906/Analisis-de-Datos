## Departamentos con mas actividad

```python
#Obteniendo las IP mas recurrentes
unicos = np.unique(data["Department"].astype(str), return_counts=True)
data_unicos = pd.DataFrame(unicos).T
data_unicos.columns=["Variable", "Conteo"]
data_unicos = data_unicos.sort_values(by=["Conteo"], ascending=False)
data_unicos.head(15) #Muestra los primeros 5 registros obtenidos


#TOP 15 de las mas usadas
top = data_unicos[0:15]["Variable"] # Top 20 "Department" a Graficar, 21 porque la categoria "nan" No la grafica
#top
```

```python
# GRAFICANDO

departamentos = top
dataset_seleccion = data[data["Department"].isin(departamentos)] #Se seleccionan los registros del top15

#Graficando barras
sns.countplot(y="Department", data=dataset_seleccion, 
              order = dataset_seleccion["Department"].value_counts().index, hue="Score")
sns.set_palette("Set2")
plt.title("Top 20 Departamentos por Score",fontsize="10")
plt.xlabel("Conteo")
plt.ylabel(None)
plt.legend(title="Score", loc="lower right", fontsize="9")
plt.grid(axis="x")
plt.show()
```

## Departamentos con mas activiad por estado

```python
# GRAFICANDO

departamentos = top
dataset_seleccion = data[data["Department"].isin(departamentos)] #Se seleccionan los registros del top15

#Graficando barras
sns.countplot(y="Department", data=dataset_seleccion, 
              order = dataset_seleccion["Department"].value_counts().index, hue="Status")
sns.set_palette("Set2")
plt.title("Top 20 Departamentos mas incidentes por Status",fontsize="10")
plt.xlabel("Conteo")
plt.ylabel(None)
plt.legend(title="Estado", loc="lower right", fontsize="9")
plt.grid(axis="x")
plt.show()
```
